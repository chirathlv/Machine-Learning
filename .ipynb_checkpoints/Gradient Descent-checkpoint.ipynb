```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Function
N = 50
X = np.linspace(-10,10, N)
J = X**2
```

$J = W^2$

$\frac{\partial J}{\partial W} = 2W, $
Set Iniial $W = 60,$
$\alpha = 0.1$

$Formula: W \leftarrow W -\alpha*\frac{\partial J}{\partial W}$

$Iteration 1: W \leftarrow 60 -0.1*2*60 = 48$

$Iteration 2: W \leftarrow 48 -0.1*2*48 = 38.4$

$Iteration 3: W \leftarrow 38.4 -0.1*2*38.4 = 30.72$

```python
# plot the graph
plt.plot(X,J)
```

```python
# Update Rule
def update_rule(w, lr):
    w = w - lr*2*w
    return w
```

```python
# Gradient Descent
def Gradient_Descent(w, lr, iterations):
    final = []
    final.append(w)
    for i in range(iterations):
        w = update_rule(w, lr)
        final.append(w)
    return final
```

```python
# Initial Parameters
initial_W = 60
learning_rate = 0.1
num_iterations = 29 #(This does not include the Initial_W value)

# Applying Gradient Descent
W = Gradient_Descent(initial_W, learning_rate, num_iterations)
Iterations = np.arange(0,30)
```

```python
# Plottin Gradient Descent
plt.figure(figsize=(15,10))
plt.plot(Iterations, W, linewidth=3, color="red")
plt.scatter(Iterations, W, linewidth=3, color="black")
plt.xlabel("Num of Iterations", fontsize=15)
plt.ylabel("W", fontsize=15)
for i,j in zip(Iterations,W):
    plt.annotate(str(j)[:3],xy=(i,j), fontsize=15)
plt.show()
```

## Applying Gradient Descent for 2 Parameters

$ J = W_1^2 + W_2^4$

```python
W1 = np.arange(1,10) # W1 parameter
W2 = np.arange(1,10) # W2 parameter
Jw = W1**2 + W2**4 # Objective Function
```

```python
# Update Rule Func
def updateRule(w1,w2,lr):
    w1 = w1 - lr*2*w1
    w2 = w2 - lr*4*w2**3
    return w1,w2
```

```python
# Apply Gradient Descent
def applyGD(w1, w2, lr, iterations):
    results1 = []
    results2 = []
    results1.append(w1)
    results2.append(w2)
    for i in range(iterations):
        w1, w2 = updateRule(w1, w2, lr)
        results1.append(w1)
        results2.append(w2)
        print(w1, w2)
```

```python
# Initial parameters
initial_w1 = 10
initial_w2 = 500
lr = 0.0000001
iterations = 1000

applyGD(initial_w1, initial_w2, lr, iterations)
```

## This can Quite after 

#### 1. Predetermined number of steps (iterations) or
#### 2. When change in W is smaller than a predetermined threshold

If Learning Rate is

<b>1. Too Big then Won't Converge (Will bounce back and forth accross the "Valley")</b>

<b>2. Too Small then Gradient Descent will be slow</b>
