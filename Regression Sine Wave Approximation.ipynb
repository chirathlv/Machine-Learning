# Sine Wave Approximation

```python
# import all the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
# Sine wave data creation
X = np.linspace(0, np.pi*8, 100)
y = np.sin(X)
```

```python
# Plot the wave
plt.figure(figsize=(10,5))
plt.plot(X,y)
plt.title("Sine Wave")
plt.show()
```

```python
# Add Polynomial Columns
def MakePoly(X, degree):
    size = X.shape[0]
    X_poly = np.ones(size).reshape(1,size).T
    
    for i in range(1, degree+1):
        P = (X**i).reshape(1,size).T
        X_poly = np.concatenate((X_poly, P), axis=1)
        
    return X_poly
```

```python
# Return the Weights
def fit(X,y):
    return np.linalg.solve(np.dot(X.T,X), np.dot(X.T,y))
```

```python
# plot each aproximated polynomial based on samples (Training Data)
def plot_graph(X, y, sample, deg):
    
    train_ids = np.random.choice(X.shape[0], sample)
    X_train, y_train = X[train_ids], y[train_ids]
    
    X_train_poly = MakePoly(X_train, deg)
    W = fit(X_train_poly, y_train)
    
    X_poly = MakePoly(X, deg)
    Yhat = np.dot(X_poly, W)
    
    plt.figure(figsize=(10,10))
    plt.plot(X, y)
    plt.plot(X, Yhat)
    plt.scatter(X_train, y_train)
    plt.xlim([0,30])
    plt.ylim([-5,5])
    plt.legend(["True","Approx"])
    plt.show()
```

```python
# Samples = 10, Degrees = 10
plot_graph(X,y,10,10)
```

```python
# Samples = 10, Degrees = 20
plot_graph(X,y,10,20)
```
