```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create an Upward Trend
num_points = 50
X = np.linspace(0,10,50)
y = X*0.5 + np.random.randn(num_points) # Adding a random noise
```

```python
# Plot the data
plt.scatter(X,y)
plt.ylim([-3,10])
plt.show()
```

```python
# Add Two Outliers
y[-1] += 30
y[-2] += 30
```

```python
# plot after adding outliers
plt.scatter(X,y)
plt.show()
```

```python
# Add Bias Term
X = np.vstack((np.ones(X.shape[0]), X)).T
```

```python
# Calculate the w
w = np.linalg.solve(np.dot(X.T,X), np.dot(X.T,y))
```

```python
# Predict the outcome
Yhat = np.dot(X, w)
```

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.scatter(X[:,1],y)
plt.plot(X[:,1],Yhat, color="red", linewidth=3)
plt.show()
```

As we can see Regression line had moved towrds the outliers. So, can see it had been affected by the outliers

```python
# Apply Ridge Regression
```

```python
# Re-calculate the w
lam = 1000
w_ridge = np.linalg.solve(np.dot(X.T,X)+lam*np.eye(X.shape[1],X.shape[1]), np.dot(X.T,y))
print(w_ridge)
```

```python
# new prediction
Yhat_ridge = np.dot(X, w_ridge)
```

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.scatter(X[:,1],y)
plt.plot(X[:,1],Yhat, color="red", linewidth=3, label="Maximum Likelihood")
plt.plot(X[:,1],Yhat_ridge, color="orange", linewidth=3, label="Maximum a Posterior")
plt.legend()
plt.show()
```

After adding $\lambda||w||^2 $ to the Cost function final weights are more Generalized than before
